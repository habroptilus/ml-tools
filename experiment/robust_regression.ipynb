# Tukey回帰

```python
import sys,os
sys.path.append(os.pardir)
from tools.data_generator.anomaly_data_generator import AnomalyDataGenerator
from tools.utils.plot.linear_plot import LinearPlotter
from tools.models.regressors.tukey_regression import TukeyRegression
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
N = 50
seed = 10
theta=(1,0.2)
```

```python
gen = AnomalyDataGenerator(theta1=1,theta2=0.2,seed=seed)
x, y = gen.run(N)
```

```python
p = LinearPlotter()
p.plot_data(x, y, theta=theta)
```

```python
model = TukeyRegression(seed=seed)
model.fit(x, y)
print(model.theta)
p.plot_model(x, y,model.theta)
```

```python
model = LinearRegression()
model.fit(np.reshape(x,(len(x),1)), y)
theta1=model.coef_[0]
theta2=model.intercept_
print(theta1,theta2)
p.plot_model(x, y,(theta1,theta2))
```
