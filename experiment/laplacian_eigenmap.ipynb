# ラプラス固有写像の実験

```python
import sys,os
sys.path.append(os.pardir)
from tools.models.decomposition.laplacian_eigenmap import LaplacianEigenmap
import numpy as np
from numpy.random import rand
```

```python
def generate(n,seed=None):
    a = 3*np.pi*rand(n)
    X_T=np.array([a*np.cos(a),30*rand(n),a*np.sin(a)])
    return X_T.T , a
```

```python
data,a=generate(1000)
```

```python
data.shape,a.shape
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
def plot3D(data,c,title):
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    ax.scatter(data[:,0],data[:,1],data[:,2],marker="o",linestyle='None',c=c, cmap='rainbow')
    plt.title(title)
    plt.show()

def plot2D(data,c,title):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.scatter(data[:,0],data[:,1],marker="o",linestyle='None',c=c, cmap='rainbow')
    plt.title(title)
    plt.show()
```

```python
plot3D(data,a,"Generated")
```

```python
model=LaplacianEigenmap(n_components=2,k=5)
```

```python
transformed=model.transform(data)
```

```python
transformed.shape
```

```python
plot2D(transformed,a,"Result")
```
