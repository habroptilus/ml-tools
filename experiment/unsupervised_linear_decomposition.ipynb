# 教師なし線形次元削減

```python
from numpy.random import randn,rand
import numpy as np
import matplotlib.pyplot as plt
import sys,os
sys.path.append(os.pardir)
from tools.models.decomposition.lpp import LPP
from tools.models.decomposition.pca import PCA
from tools.utils.plot.unlabeled_data_plot import UnlabeledDataPlotter
```

```python
def generate(n,seed=None):
    np.random.seed=seed
    xa1=2*randn(n)
    xa2=randn(n)
    xb1=2*randn(n)
    xb2=2*np.round(rand(n))-1+ randn(n)/3
    x_a=np.array([xa1,xa2])
    x_b=np.array([xb1,xb2])
    return x_a.T,x_b.T
```

```python
N=100
SEED=1
```

```python
X_a,X_b=generate(N,SEED)
```

```python
plotter=UnlabeledDataPlotter()
```

```python
plotter.plot(X_a,"generated(A type)")
```

```python
plotter.plot(X_b,"generated(B type)")
```

## PCA

```python
pca=PCA(n_components=1)
```

```python
pca_comp_a=pca.fit(X_a)[0]
pca_comp_b=pca.fit(X_b)[0]
```

```python
theta_pca_a=(0,1/pca_comp_a[0],-1/pca_comp_a[1])
theta_pca_b=(0,1/pca_comp_b[0],-1/pca_comp_b[1])
```

```python
plotter.plot(X_a,title="Result (PCA, A type)",theta=theta_pca_a)
```

```python
plotter.plot(X_b,title="Result (PCA,B type)",theta=theta_pca_b)
```

## LPP

```python
lpp=LPP(n_components=1)
```

```python
lpp_comp_a=lpp.fit(X_a)[0]
lpp_comp_b=lpp.fit(X_b)[0]
```

```python
theta_lpp_a=(0,1/lpp_comp_a[0],-1/lpp_comp_a[1])
theta_lpp_b=(0,1/lpp_comp_b[0],-1/lpp_comp_b[1])
```

```python
plotter.plot(X_a,title="Result (LPP, A type)",theta=theta_lpp_a)
```

```python
plotter.plot(X_b,title="Result (LPP, B type)",theta=theta_lpp_b)
```
