```python
import numpy as np
import sys,os
sys.path.append(os.pardir)
from tools.utils.plot.cluster_plot import ClusterPlotter
from tools.data_generator.clustered_data_generator import ClusteredDataGenerator
from tools.models.clustering.k_means import KMeans
```

```python
N = 250 # 一つのクラスに属するサンプルの個数 
mus=np.array([[2,2],[-2,-2],[-2,2],[2,-2]])
sigmas=np.array([np.eye(2) for _ in range(mus.shape[0])])
```

```python
gen=ClusteredDataGenerator(mus=mus,sigmas=sigmas)
```

```python
X,y=gen.run(N)
```

```python
plotter=ClusterPlotter()
```

```python
plotter.plot(X,y)
```

```python
model=KMeans(class_num=mus.shape[0])
```

```python
clustered=model.fit(X)
```

```python
plotter.plot(X,clustered)
```
